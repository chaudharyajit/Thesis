```python
import pandas as pd
import re

# Define the function to clean text
def clean_text(text):
    if isinstance(text, str):  # Check if the input is a string
        # Remove HTML tags
        text = re.sub(r'<[^>]+>', '', text)
        # Remove special characters but keep Devanagari characters, digits, spaces, and newlines
        text = re.sub(r'[!0-9"#$%&\'—()–*+,-./:;<=>?@[\]^_`{|}~।॥०-९]', '', text)
        # Normalize white space
        text = re.sub(r'\s+', ' ', text).strip()
    return text

# Apply the function to each cell in the specified columns
df_mapped['Content'] = df_mapped['Content'].apply(clean_text)
df_mapped['Category'] = df_mapped['Category'].apply(clean_text)

# Print the last few rows to verify
print(df_mapped.tail())


# # Save the cleaned DataFrame to a new CSV file
df_mapped.to_csv('/content/drive/MyDrive/data/Multi-Class/cleaned_output.csv', index=False)
```

```python
import pandas as pd

# Define the path to the stopwords file
stopwords_file_path = '/content/drive/MyDrive/data/Multi-Class/stopword.txt'

# Read stopwords
with open(stopwords_file_path, 'r', encoding='utf-8') as file:
    stopwords = set(file.read().splitlines())

# Define the path to the cleaned output CSV
input_csv_path = '/content/drive/MyDrive/data/Multi-Class/cleaned_output.csv'

# Read the CSV file
mapped = pd.read_csv(input_csv_path)

# Function to remove exactly matched words
def remove_exact_match_words(text, stopwords):
    if isinstance(text, str):
        return ' '.join([word for word in text.split() if word not in stopwords])
    return text

# Remove exactly matched stopwords from the 'Content' column
df_mapped = mapped.copy()
df_mapped['Content'] = df_mapped['Content'].apply(remove_exact_match_words, stopwords=stopwords)

# Save the cleaned DataFrame to a new CSV file
output_csv_path = '/content/drive/MyDrive/data/Multi-Class/stopwords_removed.csv'
df_mapped.to_csv(output_csv_path, index=False)

print(df_mapped.tail())  # Verify the result
```

### Create balance dataset

```python
import pandas as pd
from sklearn.utils import resample

# Load the CSV file
file_path = '/content/drive/MyDrive/data/Multi-Class/stopwords_removed.csv'
df = pd.read_csv(file_path)

# Display class distribution
print("Class distribution before downsampling:")
print(df['Category'].value_counts())

# Identify the minimum class size
min_class_size = df['Category'].value_counts().min()

# Downsample each class to the minimum class size
balanced_dfs = []
for category in df['Category'].unique():
    category_df = df[df['Category'] == category]
    downsampled_category_df = resample(
        category_df,
        replace=False,  # Do not replace samples
        n_samples=min_class_size,  # Match the smallest class size
        random_state=42  # For reproducibility
    )
    balanced_dfs.append(downsampled_category_df)

# Combine all downsampled classes into a single DataFrame
balanced_df = pd.concat(balanced_dfs)

# Shuffle the resulting balanced dataset
balanced_df = balanced_df.sample(frac=1, random_state=42).reset_index(drop=True)

# Display class distribution after downsampling
print("Class distribution after downsampling:")
print(balanced_df['Category'].value_counts())

# Save the balanced dataset to a new CSV file
output_path = '/content/drive/MyDrive/data/Multi-Class/balanced_dataset.csv'
balanced_df.to_csv(output_path, index=False)

print(f"Balanced dataset saved to: {output_path}")
```
